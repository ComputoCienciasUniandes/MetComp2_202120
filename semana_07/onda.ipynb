```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n_points = 1000
x = np.linspace(0.0, 1.0, n_points)
#u_inicial = np.exp(-0.5*((x-0.3)/(0.1))**2)
u_inicial = np.sin(x*2.0*np.pi)
```

```python
plt.plot(x, u_inicial)
```

```python
delta_x = x[1] - x[0]
delta_t = 0.0005
c = 1.0
r = c * delta_t / delta_x
print(r)
```

```python
u_futuro = u_inicial.copy()
for i in range(1,n_points-1):
    u_futuro[i] = u_inicial[i] + (r**2/2.0) * (u_inicial[i+1] - 2.0* u_inicial[i]+ u_inicial[i-1])
```

```python
# n+1 -> futuro
# n -> presente
# n-1 -> pasado
```

```python
u_pasado = u_inicial.copy() # n=0
u_presente = u_futuro.copy() # n=1
```

```python
n_time = 700
for j in range(n_time):
    for i in range(1,n_points-1):
        u_futuro[i] = 2.0*(1-r**2)*u_presente[i] - u_pasado[i] + (r**2)*(u_presente[i+1]+ u_presente[i-1])
    
    u_pasado = u_presente.copy()
    u_presente = u_futuro.copy()
        
        
```

```python
plt.plot(x, u_inicial)
plt.plot(x, u_futuro)
```
