```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def func_prime_1(x, y):
    return -np.sin(x)
```

```python
def euler(x_old, y_old, h):
    x_new = x_old + h
    k = func_prime_1(x_old, y_old)
    y_new = y_old + h * k
    return x_new, y_new
```

```python
def RK4(x_old, y_old, h):
    k_1 = func_prime_1(x_old, y_old)
    
    x1 = x_old + 0.5*h
    y1 = y_old + 0.5*h*k_1
    k_2 = func_prime_1(x1, y1)
    
    x2 = x_old + 0.5*h
    y2 = y_old + 0.5*h*k_2
    k_3 = func_prime_1(x2, y2)
    
    x3 = x_old + h
    y3 = y_old + h*k_3
    k_4 = func_prime_1(x3, y3)
    
    k = (1.0/6.0)*(k_1 + 2.0*k_2 + 2.0*k_3 + k_4)
    
    x_new = x_old + h
    y_new = y_old + h * k

    return x_new, y_new
```

```python
min_x = 0.0
max_x = 20.0*np.pi
h = 0.1
n_points = int((max_x-min_x)/h)
print(n_points)

x = np.zeros(n_points)
y = np.zeros(n_points)

x[0] = min_x
y[0] = 1.0

for i in range(1,n_points):
    x[i], y[i] = RK4(x[i-1], y[i-1], h)
```

```python
plt.plot(x,y)
plt.plot(x, np.cos(x))
```
