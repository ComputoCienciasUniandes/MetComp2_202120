```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n_x = 101
n_t = 10000

nu = 0.1
sigma = 0.01

x = np.linspace(0.0, 2.0*np.pi, n_x)
dx = x[1] - x[0]

dt = sigma * dx**2/nu
alpha = dt/dx**2

print(dx, dt, alpha)
```

```python
u = np.sin(x)
u_inicial = u.copy()
plt.plot(x, u)
```

```python
for n in range(n_t):
    u_past = u.copy()
    for i in range(1,n_x-1):
        u[i] = u_past[i]
        u[i] += -u_past[i]*(dt/dx)*(u_past[i] - u_past[i-1])
        u[i] += nu * alpha * (u_past[i+1] - 2.0*u_past[i] + u_past[i-1])
```

```python
plt.plot(x, u_inicial)
plt.plot(x, u)
```
